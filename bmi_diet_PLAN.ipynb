```python
## K-Means on categorical data

def clusterKmeans(catdata):
    X = np.array(catdata[:,1:len(catdata)])
    kmeans = KMeans(n_clusters=5, #  number of clusters to form
                    init='k-means++',# method for initializing cluster centroids
                    max_iter=300, # maximum number of iterations for algo
                    tol=0.0001,   # tolerance level for convergence
                    algorithm='elkan', #
                    random_state=0).fit(X)# fits the K-Means model to the data in the array X, and kmeans holds the fitted model
    lbl=kmeans.labels_  # extracts the cluster labels assigned by the K-Means algorithm to each data point
    return lbl
```

```python
def calBmi(weight, height): #in kg,cm
    bmi = weight/((height/100)**2) #cm to mt
    print("Your body mass index is: ", round(bmi,2))
    if ( bmi < 16):
        print("Acoording to your BMI, you are Severely Underweight")
        bfbmi=2
        lbmi=2
        dbmi=3
    elif ( bmi >= 16 and bmi < 18.5):
        print("Acoording to your BMI, you are Underweight")
        bfbmi=4
        lbmi=0
        dbmi=2
    elif ( bmi >= 18.5 and bmi < 25):
        print("Acoording to your BMI, you are Healthy")
        bfbmi=3
        lbmi=3
        dbmi=0
    elif ( bmi >= 25 and bmi < 30):
        print("Acoording to your BMI, you are Overweight")
        bfbmi=1
        lbmi=4
        dbmi=4
    elif ( bmi >=30):
        print("Acoording to your BMI, you are Severely Overweight")
        bfbmi=0
        lbmi=1
        dbmi=1
    return bfbmi, lbmi, dbmi #tuples
```

```python
#Creating TrainingDataset

def makeTrainingset(cat, lbl, bmicls):
    # Create an empty array to store the augmented feature vectors.
    catfin = np.zeros((len(cat)*5, 6), dtype=np.float32)

    # Iterate over the BMI classes and the cat list.
    for zz in range(5):
        for jj in range(len(cat)):

            # Create a new feature vector.
            valloc=list(cat[jj]) #weightloss major columns -4
            valloc.append(bmicls[zz]) #add bmi value randomly
            #valloc.append(agecls[zz]) #add age value  randomly

            # Store the new feature vector in the catfin array.
            catfin[t] = np.array(valloc)

            # Append the label to the yt list.
            yt.append(lbl[jj])

            # Increment the row index.
            t += 1

    # Return the catfin and yt lists.
    return catfin, yt
```

```python
def calculate_bmi(height, weight):
    height /= 100
    bmi = weight/np.square(height)
    return round(bmi,2)
```

```python
def get_min_max(col_name, df):
    q1 = df[col_name].quantile(0.25)
    q3 = df[col_name].quantile(0.75)
    iqr = q3-q1
    q_min = q1-(1.5*iqr)
    q_max = q3+(1.5*iqr)

    return q_min,q1,q3,q_max    
```
